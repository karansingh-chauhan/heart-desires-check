```python
train_df=pd.read_csv("D:\\machine learning projecsts folders\\hearth desese\\train.csv")
test_df=pd.read_csv("D:\\machine learning projecsts folders\\hearth desese\\test.csv")
```

changing heart diese into numerical

```python
train_df['Heart Disease']=train_df['Heart Disease'].map({'Presence':1,'Absence':0})
```

```python
train_df.drop(columns=['id'],inplace=True)
```

```python
train_df.sample(3)
```

```python
df=train_df.copy()
```

```python
X=df.drop('Heart Disease',axis=1)
y=df['Heart Disease']
```

feature selction

correlation feature importance

```python
fi_df1=df.corr()['Heart Disease'].iloc[1:].to_frame().reset_index().rename(columns={'index':'col_name','Heart Disease':'heatmap'})\
    .sort_values('heatmap',ascending=False)
fi_df1
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(X,y)
```

```python
fi_df2=pd.DataFrame({
    'col_name':X.columns,
    'radomforst':model.feature_importances_
}).sort_values('radomforst',ascending=False)
fi_df2
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model2=GradientBoostingClassifier()
model2.fit(X,y)
```

```python
fi_df3=pd.DataFrame({
    'col_name':X.columns,
    'gradientbost':model2.feature_importances_   
}).sort_values('gradientbost',ascending=False)
fi_df3
```

```python
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
model3=RandomForestClassifier(n_estimators=100,random_state=42)
model3.fit(x_train,y_train)
perm_imp=permutation_importance(model3,x_test,y_test,n_repeats=30,n_jobs=-1,random_state=42)
```

```python
fi_df4=pd.DataFrame({
    'col_name':X.columns,
    'perm_imp':perm_imp.importances_mean
}).sort_values('perm_imp',ascending=False)
fi_df4
```

linear model calssifiction

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train_1=st.fit_transform(X)
```

```python
model4=LogisticRegression()
model4.fit(x_train_1,y)
```

```python
fi_df5=pd.DataFrame({
    'col_name':X.columns,
    'logistic':model4.coef_[0]
}).sort_values('logistic',ascending=False)
fi_df5
```

```python
from sklearn.feature_selection import RFE

# Initialize the base estimator
estimator = RandomForestClassifier()

# Apply RFE on the label-encoded and standardized training data
selector_label = RFE(estimator, n_features_to_select=X.shape[1], step=1)
selector_label = selector_label.fit(X, y)

# Get the selected features based on RFE
selected_features = X.columns[selector_label.support_]

# Extract the coefficients for the selected features from the underlying linear regression model
selected_coefficients = selector_label.estimator_.feature_importances_

# Organize the results into a DataFrame
fi_df6 = pd.DataFrame({
    'col_name': selected_features,
    'rfe': selected_coefficients
}).sort_values(by='rfe', ascending=False)

fi_df6
```
